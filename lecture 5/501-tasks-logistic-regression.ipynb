## Confusion Matrix

The confusion matrix is a powerful tool for measuring the performance of a classifier. It is similar to plotting Predicted vs. Real values when using regression models but instead is a count of how many times the predicted value occurs compared to the real value. For example in a binary situation with two values `Positive` or `Negative`, the confusion matrix would look like this:

|| Predicted `Negative` | Predicted `Positive` |
|--|--|--|
|<b>Real `Negative`</b>|True Negatives| False Positives|
|<b>Real `Positive`</b> | False Negatives|True Positives|

In the field of diagnostics, this matrix contains the elements required for calculated <b>Specificity</b>, <b>Sensitivity</b>, and <b>PPV</b>. Below we construct a confusion matrix to analyse the performance of our classifier.
