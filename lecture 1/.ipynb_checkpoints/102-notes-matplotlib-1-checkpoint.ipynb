# `pandas` with `matplotlib`

Now that we are familiar with the basics of `DataFrame` objects, we can use them to manage and store data ready for plotting using `matplotlib` and specifically `matplotlib.pyplot`. We will now present a set of examples that cover some advanced usages of `pandas` and `matplotlib.pyplot`. We are also going to reintroduce `numpy` and use this for some of our calculations.

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

## Simple Example

First we will remind ourselves of how to use `np.arange` by creating some sample data. The sample data is stored in a `DataFrame` called `data`. We initialise data in a one-liner using the following command:

```python
data = pd.DataFrame({
    'x' : np.arange(0, 10, 0.1)
})
```

This is the same as creating the dictionary object first in a separate variable:

```python
data_dictionary = {
    'x' : np.arange(0, 10, 0.1),
}

data = pd.DataFrame(data_dictionary)
```

> We can use one-liners to reduce the number of different variables stored in memory, but it is important to note that sometimes this makes the code written less legible.

The column called `'x'` will be the data on the x-axis, and the next step is to create some data and add it dynamically into the `DataFrame` we have just created. In this case we will use the sine function and use the vectorised form of this function that is stored within `numpy` to work on the whole column at the same time:

```python
data['sin(x)'] = np.sin(data['x'])
```

> Vectorised functions work by acting on array-like data, by performing the same function on the data without using a for loop

Note how we did not have to manually create an empty `list` or `array` beforehand, we can just dynamically create the column called `'sin(x)'` in one line.

After this, it is time to start using `matplotlib.pyplot`. It can be seen that the package has been shortened to `plt` which is the generally accepted abbreviation for `matplotlib.pyplot`. Here are the steps needed to create a successful plot:

1. Initialise figure and axes
2. Do any pre-plotting formatting, such as adding labels
3. plot the data
4. Do any post-plotting formatting, such as adding a legend
5. Save or show figure

Throughout these notes, we will endeavour to ensure that this format is followed, to make the code more legible and easier to follow. It is important however to be aware that this format is not followed by everyone in the scientific community and for some examples it is too simple for the application. As we move through the course, we will point out when this scheme for organising the code is not used.

To create an axis, we generally use the `plt.subplots` function. This creates both `plt.Figure` and `plt.Axes` objects and links them together. In some cases, such as for 3D plots, this will not be used, but in general it makes a good starting point for making a plot and indicating to someone who is reading your code that you are about to begin plotting. The `plt.subplots` function is called like this:

```python
fig, ax = plt.subplots()
```

The object called `ax` is the main object used for plotting, it contains a range of functions that allow for customisation of the axes, any data plotted and any annotations to be drawn. The next step is to do some rudimentary formatting of the axes by adding labels to them. In this case we use the `plt.Axes.set_xlabel` and `plt.Axes.set_ylabel` <b>methods</b> (remember that methods belong to classes) to do this:

```python
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
```

> If you have used `matplotlib.pyplot` before, you may have noticed people using `plt.xlabel` or `plt.ylabel` functions. These functions add labels to the most recent `plt.Axes` object created which can sometimes have unexpected results. This is why we prefer to use `plt.Axes.set_xlabel` and `plt.Axes.set_ylabel`

Now we get to doing the plotting, for which there are many different methods that can be used. In this case, we will use one of the most common methods called `plt.Axes.plot`. This is a smart function that can take a variety of arguments and keyword arguments in different combinations. In this case, we will use it in the following way:

- Argument 1: x-data
- Argument 2: y-data
- Argument 3: format string

The format string describes the appearance of the line and is a shortcut for simple preset formats. For more complicated formats, it is better to use keyword arguments. We will show examples of this later in the course. For now the format string will be `'kx-'` where `'k'` refers to colour black, `'x'` uses a cross for each point, and `'-'` means that a solid line will be used. Altogether this looks like:

```python
ax.plot(data['x'], data['sin(x)'], 'kx-')
```

Finally the `plt.show` function is used to present the graph, although this is not always necessary in Jupyter Notebook. Now that we have walked through each of the steps, we will continue with more examples and highlight how they are different to this template.

> When scripting using <b>.py</b> files, we can also use the `fig.show` function but this raises a warning when using Jupyter Notebook

````python
# create our independent variable x
data = pd.DataFrame({
    'x' : np.arange(0, 10, 0.1)
})

# create the dependent variable sin(x)
data['sin(x)'] = np.sin(data['x'])
 
# initialise axes
fig, ax = plt.subplots()

# adjust settings
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')

# call plot function
ax.plot(data['x'], data['sin(x)'], 'kx-')

# show
plt.show()
````

## Multiple Data and Adding a Legend

It is more likely that more than one data source will be plotted on the same axes, so the following example demonstrates a smart way of doing this and also adding a legend.

### Adding the legend

Working backwards through the code, we now see that after the plotting stage of the code, there is a new line:

```python
ax.legend()
```

The `plt.Axes.legend` method adds a legend to the plot. How the names are determined will be explained now.

### Using `DataFrame` column names as data labels

To provide a legend, a name in the form of a `str` object must be provided each time the `plt.Axes.plot` method (or an equivalent plotting method) is called. This is done by using the `label` keyword argument. This is all done within the following loop:

```python
for key, value in data.items():
    if key != 'x':
        ax.plot(data['x'], value, label=key)
```

When inspecting the loop, it is clear to see that it takes a similar format to looping through a dictionary, because the `pd.DataFrame.items` method is used to create key:value pairs. In this case the `key` variable is passed to the `label` keyword argument by using `label=key` at the end of the `ax.plot` method. The value is the data stored in that column. If we were to add a statement `print(key)` within the loop then the output would look like this:

```python
'x'
'sin(x)'
'cos(x)'
```

We also add an `if` statement which just ensures that a plot of `'x'` vs. `'x'` is not plotted.

````python
# create our independent variable x
data = pd.DataFrame({
    'x' : np.arange(0, 10, 0.1)
})

# create the dependent variable sin(x)
data['sin(x)'] = np.sin(data['x'])
data['cos(x)'] = np.cos(data['x'])

# initialise axes
fig, ax = plt.subplots()

# adjust settings
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

# create loop and call plot function
for key, value in data.items():
    if key != 'x':
        ax.plot(data['x'], value, label=key)
        
# any more settings
plt.legend()

# show
plt.show()
````

## LaTeX labelling, fontsize and logarithmic axes


### Setting fontsize and tidying legend

One thing that is particularly important when producing plots for publications or report is the need for them to be legible and presented tidily. In general, scientific conventions state that plots should contain the least amount of embellishment necessary to effectively communicate the data.

The fontsize of all the text is too small, and this effect is more noticeable when a file is exported and added to a document. To fix this, we add the keyword argument `fontsize={size}` where `{size}` can take a variety of values, that are described [here](). In our example, the size used is passed as a string: `'xx-large'`. The final result is this:

```python
ax.set_xlabel("$x$", fontsize='xx-large')
ax.set_ylabel("$f(x)$", fontsize='xx-large')
```

Further down in the code, there are two `for` loops that increase the size of the labels marking out the values of the ticks. These snippets look like this for the x- and y-axis respectively. The important thing to note here is that every feature of a plot created by `matplotlib` is some kind of Python object, and that they can be accessible, sometimes in unexpected ways.

```python
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize('xx-large')
    
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize('xx-large')
```

The legend has a box around it (this is a smaller point but still necessary to produce a graph that is as tidy as possible). To remove the box we pass the `frameon=False` keyword through to the `plt.Axes.legend` method. Additionally, the fontsizes of the text in the legend must be increased as well.

```python
ax.legend(frameon=False, fontsize='xx-large')
```

### Latex string formatting

In scientific computing, a typesetting language call LaTeX is one of the main ways for producing reports. One of the reasons for this is power of mathematical formatting to create equations and other mathematical representations. This has been translated to many different frameworks such as Javascript, Markdown (the typesetting used for making these notes) and Python. Within Python, we can make strings called <b>raw string literals</b> which allow `matplotlib` to correctly format things like superscripts, equations, summations, integrals using the LaTeX common typesetting format. We can do this in a smart way by setting the column names in our `DataFrame` object to these correctly formatted raw string literals by wrapping the strings in dollar signs, so they look like this: `"$<math formatting>$"`. This might sound confusing, but in practice is simple, and looks like this:

```python
data["e^x"] = np.exp(data['x'])
data["e^{2x}"] = np.exp(2 * data['x'])
...
for key, value in data.items():
    if key != 'x':
        ax.plot(data['x'], value, label=f"${key}$")
```

> Formatting using LaTeX math strings is quite complicated to understand at first but there is a wealth of information online

### Logarithmic Axes

Finally, since our functions are exponentials, it makes sense to present them on logarithmic axes. To do this is simple, and we used a `plt.Axes.set_` method that looks like this:

```python
ax.set_yscale('log')
```

````python
# create our independent variable x
data = pd.DataFrame({
    'x' : np.arange(0, 10, 0.1)
})

# create the dependent variables - use latex/markdown formatting to create proper representation of powers
data["e^x"] = np.exp(data['x'])
data["e^{2x}"] = np.exp(2 * data['x'])

# initialise axes
fig, ax = plt.subplots()

# adjust settings
ax.set_xlabel("$x$", fontsize='xx-large')
ax.set_ylabel("$f(x)$", fontsize='xx-large')
ax.set_yscale('log')

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize('xx-large')
    
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize('xx-large')

# create loop and call plot function
for key, value in data.items():
    if key != 'x':
        ax.plot(data['x'], value, label=f"${key}$")

ax.legend(frameon=False, fontsize='xx-large')
plt.show()
````

## Using plotting functions, other plot types, and subplots

The final plot of this part of the course looks like it would be very complicated to plot since there are so many features. We can simplify the plotting process in the same way that we simplify many other things in Python: by utilising functions. This means that when it comes to plotting our graph, it only takes 3 lines (assuming that we have already generated our data):

```python
fig, ax = plt.subplots(ncols=2)
plot_histograms(ax[0], data)
plot_power(ax[1], data)
```

In the final section we will go through each of these functions and describe how to use functions to plot data on existing `plt.Axes` by remembering that they are just Python objects that can be passed as arguments to functions.

> Condensing segments of code with many lines, e.g. for plotting, down to functions that can be called with one line is highly desirable and makes the reading experience optimised. A further point to add is that it means that if a function is used many times, any updates only need to be made once

### Plotting a moving average

To plot a moving average, we used the `pd.DataFrame.rolling` method. This takes one argument which is the window size, in terms of number of rows. In our function, we have this as a keyword argument, but if we didn't then the method would like this:

```python
data = {
    ...
}
df = pd.DataFrame(data)
moving_df = df.rolling(20)
```
The `df.rolling(20)` statement returns a new object that can be used to generate a moving average by using the `.mean()` method. The final form looks like this, where `rolling` is a `DataFrame` and `window` is a keyword argument: 

```python
rolling = df.rolling(window).mean()
```

We can use `rolling` as though it is a normal `DataFrame` (which it is) and plot it to the `plt.Axes` that have been passed as the first argument of the function called `plot_power`.

### Plotting a histogram

Plotting a histogram as shown below is a little tricky to do and requires some know-how. The main structure of the function is taken from [this example]() which is on the `matplotlib.pyplot` website. This website contains a myriad of examples and to learn more about `matplotlib`, this is the place to start.

#### Histogram parameters

In the function `plot_histograms` there are 3 arguments, two of which are the same as before. The final argument called `bin_width` is specifically the size of the histogram 'bin' into which the current or voltage can be divided. For example if we have `bin_width=0.2` then the bins would be: 0.0-0.2, 0.2-0.4, 0.4-0.6, ...

The first four lines of code within the function set some parameters for creating the histogram. These are as follows:
- `bin_min` - the lower bound of the first bin 
- `bin_max` - the upper bound of the final bin
- `bins` - a list of bin positions (i.e. 0.0, 0.2, 0.4, etc...) rounded to the same precision as the `bin_width`
- `width` - the width of a histogram bar when plotted on a `plt.Axes`


#### Using a dictionary to store the data
We then pass the `bins` variable to a function called `np.histogram` along with the columns from our `DataFrame`, allocating the result to a dictionary:
```python
histograms = {
    'V' : np.histogram(data['V'], bins),
    'I' : np.histogram(data['I'], bins),
}
```

#### Plotting and correctly formatting the bar charts
Now we are ready to proceed with our usual routine of formatting the `plt.Axes` and plotting the data. In this case we use `ax.bar` which creates a bar graph that is centred on the x-value of the data. We need to offset the x-values of each data point because we are plotting two pieces of data at the same x-value. This is where the `width` variable comes in, because we can use this to shift the x-values to the left or right, allowing the overall centre of the two bars to be aligned with the actual x-value.
```python
ax.bar(
    bins[:-1] + width/2, # shifted x-value
    histograms['V'][0] / len(data['V']), # y-value
    width, # bar width
    label='Voltage [V]' # keyword argument
)

ax.bar(
    bins[:-1] - width/2, 
    histograms['I'][0] / len(data['I']), 
    width, 
    label='Current [A]'
)
```

Finally, for demonstration, we have presented different examples of plotting each function on its own `plt.Figure` and then finally plotting them on two sub-axes contained in the same `plt.Figure`, by using the `plt.subplots(ncols=2)` function.

````python
data = pd.DataFrame({
    'time' : np.arange(0., 10., 0.05)
})

data['V'] = data['time'].apply(lambda x: 1 + np.sin(x ** (1./2.)) + np.random.random() / 20.)
data['I'] = data['time'].apply(lambda x: 1 + 2 * np.random.random() / 2.)
data['P'] = data['I'] ** 2 * data['V']

def plot_power(ax: plt.Axes, df: pd.DataFrame, window: int = 20):
    """Plots the rolling average of the power"""
    
    # do calculation of moving average
    rolling = df.rolling(window).mean()
    
    # set labels
    ax.set_xlabel('Time [s]')
    ax.set_ylabel('Power [W]')
    
    # do plotting
    ax.plot(df['time'], df['P'], 'b-', alpha=0.25)
    ax.plot(rolling['time'], rolling['P'], 'r-')
    
    return

def plot_histograms(ax: plt.Axes, df: pd.DataFrame, bin_width: float = 0.2):
    """Plots histograms of the current and voltage"""
    
    # create the bins
    bin_min = min(df['I'].min(), df['V'].min())
    bin_max = max(df['I'].max(), df['V'].max())
    bins = np.arange(bin_min, bin_max, bin_width) // bin_width * bin_width
    width = 0.8 * bin_width / 2
    
    # create the histograms
    histograms = {
        'V' : np.histogram(data['V'], bins),
        'I' : np.histogram(data['I'], bins),
    }
    
    # set labels
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    ax.set_xticks(bins[:-1])
    
    # do plotting
    ax.bar(
        bins[:-1] + width/2, 
        histograms['V'][0] / len(data['V']), 
        width, 
        label='Voltage [V]'
    )
    
    ax.bar(
        bins[:-1] - width/2, 
        histograms['I'][0] / len(data['I']), 
        width, 
        label='Current [A]'
    )
    
    # set axis
    ax.legend(frameon=False, loc=2)
    
    return

# test the plotting of histograms
fig, ax = plt.subplots()
plot_histograms(ax, data)

# test the plotting of the power
fig, ax = plt.subplots()
plot_power(ax, data)

# plot them together on two sub-axes
fig, ax = plt.subplots(ncols=2)
plot_histograms(ax[0], data)
plot_power(ax[1], data)
````
