```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres
period = 0.2  # Dure d'une priode
num_periods = 5  # Nombre de priodes
step = 0.02  # Pas entre les points

# Crer une priode simple pour la partie gauche et droite
x_half = np.arange(-period/2, 0, step)
y_half = np.ones_like(x_half)
x_half2 = np.arange(0, period/2, step)
y_half2 = -np.ones_like(x_half2)

# Combiner les deux moitis d'une priode
x_single_period = np.concatenate([x_half, x_half2])
y_single_period = np.concatenate([y_half, y_half2])

# Rpter sur plusieurs priodes
x = np.tile(x_single_period, num_periods) + np.repeat(np.arange(num_periods) * period, len(x_single_period))
y = np.tile(y_single_period, num_periods)

# Tracer le signal priodique
fig, ax = plt.subplots()
ax.plot(x, y, 'k-', label='Signal priodique')

# Configuration de l'axe
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
plt.show()
```
