```python
# import matplotlib, numpy and pandas using their abbreviations
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create two variables x from 0 to 1, with a step of 0.1, and y = sin(x)
x = np.arange(0, 1, 0.1)
y = np.sin(x)
```

```python
# create a new set of axes, label the x-axis 'x' and the y-axis 'sin(x)'
fig, ax = plt.subplots()
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')

# plot y vs x using a solid black line
ax.plot(x, y, 'k-')
None
```

```python
# use latex formatting to plot and label y = x^2, using a blue dotted line
fig, ax = plt.subplots()
x = np.arange(0, 1, 0.1)
y = x ** 2
ax.plot(x, y, "b-")
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$x^2$")
None
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 1, 0.1)
y1 = np.random.rand(len(x))
y2 = np.random.rand(len(x))

# Create a figure with two subplots side by side
fig, (ax, bx) = plt.subplots(1, 2)

# Plot on the first subplot
label = r"$y_1$"
ax.plot(x, y1, 'rs', label=label)  # 'rs' means red squares
ax.legend()

# Plot on the second subplot
label = r"$y_2$"
bx.plot(x, y2, "bo", label=label)  # 'bo' means blue circles
bx.legend()

# Show the plot
plt.show()
```

```python
# Create a 2x2 set of axes
fig, ax = plt.subplots(2,2, figsize=(9, 9))

# Use numpy to create 4 different sets of data, stored in a list
data = []
for i in range(4):
    if i == 0:
        x = np.random.rand(100)
    elif i == 1:
        x = np.random.rand(100)
    elif i == 2:
        x = np.random.rand(100)
    else:
        x = np.random.rand(100)
    
    data.append(x)
# print(len(data))
# for each axes, plot a new combination of data, using a different style
ax[0][0].plot(data[2], data[1], 'k-')

ax[0][1].plot(data[3], data[2], 'bo')

ax[1][0].plot(data[0], data[3], 'rs')

ax[1][1].plot(data[2], data[0], "b-")

# set the top left axis to have a log-scale on the y-axis
ax[0][0].set_yscale('log')

None
```
